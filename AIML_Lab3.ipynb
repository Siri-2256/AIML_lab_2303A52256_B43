```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['Siri','bingo','Meenakshi','Nandi','Chandana']
age=[25,30,30,33,23]
height=[145,55,65,222,173]
weight=[45,55,63,65,92]
data={
     'Name':people,
     'Age':age,
     'Height':height,
     'Weight' : weight

}
df=pd.DataFrame(data)


plt.figure(figsize=(10,6))
plt.bar(df['Name'],df['Height'],color='blue')
plt.title('Bar chart of individual height')
plt.xlabel('Name')
plt.ylabel('Height(cm)')


plt.show()







```
